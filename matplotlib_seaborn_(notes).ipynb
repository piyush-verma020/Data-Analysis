### Styling lines and markers

The **plt.plot** function supports many argument for styling lines and markers:
* **colour** or **c**: set the colour of the line (https://matplotlib.org/stable/gallery/color/named_colors.html)
* **linestyle** or **ls**: choose between solid or dashed line
* **linewidth** or **lw**: set the width of the line
* **markersize** or **ms**: set the size of the marker
* **markeredgecolour** or **mec**: set the edge colour for markers
* **markerdgewidth** or **mew**: set the edge width for markers
* **markerfacecolour** or **mfc**: set the fill colours for markers
* **alpha**: opacity of the plot


```python
plt.xlabel('Years')
plt.ylabel('Yield (tonnes per hectare')

plt.title('Crop Yields in Canto')  # "Canto" is an imaginary place somewhere in this cosmos

plt.plot(years, apple, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, orange, marker='o', c='r', ls='--', lw=3, ms=10, mec='black',mew=3, alpha=.5)

plt.legend(['Apples', 'Orange'])
plt.show()
```

The **fmt** argument provides a shorthand for specifying the line style, marker and line colour It can be provide as the third argument to **plt.plot**.

The format of the **fmt** argument is: 

**fmt = '[marker][line_style][colour]'**

#### Note:

If no line style is specified in **fmt** argument, then **fmt** will draw only markers.

### Adding Hues

#### What is hue?

* hue tells Seaborn to separate your data into categories based on a column's values.

* It automatically assigns different colors to each unique value in that column.

#### How to use hue:

##### sns.scatterplot(data=flower_df, x='sepal_length', y='sepal_width', hue='species');

* We are plotting sepal length vs. sepal width.

* The column species has 3 unique values: 'setosa', 'versicolor', 'virginica'.

* hue='species' makes Seaborn:
    * Group the points by species

    * Assign a different color to each species

    * Add a legend so you can understand which color represents which species}
 
#### Visual Effect:
So in our scatterplot, we see:

* Red dots for Setosa

* Blue dots for Versicolor

* Green dots for Virginica
(or other colors — depending on the theme)

Each group is still plotted on the same X-Y space, but with its own color.

```python
sns.scatterplot(data=flower_df, x='sepal_length', y='sepal_width', hue='species');
```

We could also increase the size of the dots by using s = 100 as another argument like:

```python
sns.scatterplot(data=flower_df, x='sepal_length', y='sepal_width', hue='species',s=75);
```

It looks like adding hue makes the plot more informative. we can immediately tell thatthe flower of Setosa species have a smaller sepal length but higher
sepal width, while the opposite is true for the flower of Virginica species.

### Bars Plots with Averages

Let's look at another dataset included in Seaborn, called **'tips'**. This dataset contains information such as the sex of the customer,
time of day, total bill, and tip amount for customers visiting a restaurant over the course of a week.

```python
tips_df = sns.load_dataset('tips');
```

```python
tips_df
```

We might want to plot a bar chart to visualize how the average bill amout varies across different days of the week.
One way to do this would be to compute the day-wise averages and the use **plt.bar**.

However this is a very common use case, the seaborn library provides a **barplot** function which can automatically compute averages.

##### Seaborn changed the format slightly in more recent versions, and now it requires that x and y be passed as keyword arguments (instead of positional ones)

sns.barplot('day','total_bill',data=tips_df);

instead of using the above code use this instead:

**sns.barplot(x='day',y='total_bill',data=tips_df);**

```python
sns.barplot(x='day',y='total_bill',data=tips_df);
```

The lines cutting each bar represents the amount of variation in the values. For instance it seems like the variation in the total bill was quite high on 
Friday and lower on Saturday.

```python
tips_df.groupby('day')[['total_bill']].mean()
```

**FutureWarning: The default of observed=False is deprecated and will be changed to True in a future version of pandas.**
**Pass observed=False to retain current behavior or observed=True to adopt the future default and silence this warning.**

**tips_df.groupby('day')[['total_bill']].mean()**

This is not an error, just a warning from pandas.

It’s saying that in future versions, the default behavior of groupby() will change slightly.

Right now, it shows all possible categories (like all 7 days of the week), even if some (like Monday, Tuesday, and Wednesday) are not in the dataset.

In the future, it will only show the days that actually appear in your data (Thursday, Friday, Saturday, Sunday — in the case of the tips dataset).

To avoid the warning and be future-proof, you can write:

**tips_df.groupby('day', observed=True)[['total_bill']].mean()**

```python
bill_avg_df = tips_df.groupby('day', observed=True)[['total_bill']].mean() # This time no future errors will be seen
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = bill_avg_df); # Well you could just use plt.bar(bill_avg_df.day, bill_avg_df.total_bill)
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips_df);
```

We can also specify a **hue** argument to compare bar plots side-by-side based on third feature example **sex**.

```python
bill_avg_df = tips_df.groupby(['day', 'sex'])[['total_bill']].mean().reset_index()
sns.barplot(x = 'day', y = 'total_bill',hue='sex', data = bill_avg_df);
```

You could make the bars horizontal just by switching the axes (the values of x and y).

```python
sns.barplot(y = 'day', x = 'total_bill',hue='sex', data = bill_avg_df);
```

```python
sns.barplot(x = 'day', y = 'total_bill',hue='smoker', data = tips_df);
```

This is how you can use multiple features with the **hue** parameter to compare plots.

However, you cannot pass a list to **hue**, hoping it will automatically perform the comparison for you.

```python
tips_df['smoker_sex'] = tips_df['smoker'].astype(str) + ' ' + tips_df['sex'].astype(str)
```

```python
sns.barplot(x='day', y='total_bill', hue='smoker_sex', data=tips_df)
```

### Heatmap

A heatmap is used to visualize 2-Dimentional data just like a matrix or a table using colors.
The best way to understand is by looking at an example.
We will be using another sample dataset from Seaborn called **'flights'**, to visualize monthly passenger footfall at an airport over 12 years

```python
flights_df = sns.load_dataset('flights')
```

```python
pivoted_df = flights_df.pivot(index='month', columns='year', values='passengers')
```

The **.pivot()** function in pandas is used to reshape a DataFrame — it allows us to turn unique values from one column into column headers and reorganize your data into a more readable table/matrix format.

##### Example:
Suppose you have a DataFrame like this:

**pivoted_df = flights_df.pivot(index='month', columns='year', values='passengers')**

Now each month is a row, each year is a column, and each cell contains passenger numbers.

##### Points to be noted:

* The combination of index and columns must be unique. If not, use .pivot_table() instead.

* Use .pivot() when you're reshaping clean, non-repeating data.

```python
pivoted_df
```

#### **Difference between .pivot() and .pivot_table():**

#### .pivot():

* Simple reshaping

* Requires unique combinations of index and columns

* Will throw an error if there are duplicates

**Syntax:**

**pivoted_df = df.pivot(index='month', columns='year', values='passengers')**

#### Note:

**This only works if there’s exactly one value per month-year pair.**

---

#### .pivot_table():

* More flexible and powerful

* Can handle duplicate entries

* Allows aggregation (mean, sum, etc.) over duplicates

**Syantax:**

**pivot_table_df = df.pivot_table(index='month', columns='year', values='passengers', aggfunc='mean')**

#### Note:

**ven if a month-year combination appears multiple times, it will take the mean (or any function you choose) of those values.**

---

**pivoted_df** is a matrix with one row for each month and one column for each year. The values in the matrix show the number of passengers (in thousands)
that visited the airport in a specific month of a specific year.

We can use the **sns.heatmap** function to visualize the footfall at the airport.

```python
plt.title("No. of Passengers (in 1000's)")
sns.heatmap(pivoted_df);
```

The brighter color indicates higher footfall at the airport. By looking at the graph we can infer two things:

1. The footfall at the airport at any year tends to be the highest around July and August.
2. The footfall at the airport at any given month tends to grow year by year.

We can display the actual values in each block by specifying **annot=True**, and use the **cmap** argument to change the color palette.

```python
plt.title("No. of Passengers (in 1000's)")
sns.heatmap(pivoted_df, fmt = 'd', annot=True, cmap='Blues');
```

#### What is the purpose of fmt='d' in sns.heatmap()?

* fmt stands for "format" — it controls how the numbers are displayed inside each cell when annot=True is used.

* fmt='d' means: format the annotations as decimal integers (whole numbers).

So, if the numbers in your heatmap are integers like 150, 200, 300, etc., this will display them exactly as-is.

#### Why do you see values like 1e+02, 2e+02, etc., when fmt is not provided?

That’s because by default:

* matplotlib tries to automatically format numbers, and

* If values are large (e.g., in hundreds or thousands), it may switch to scientific notation like 1e+02 = 100, 2e+02 = 200.

#### Most Common Use Cases for fmt in sns.heatmap()

How to avoid scientific notation or control number formatting:

* fmt='d' → for integers (e.g., 150, 200, 300)

* fmt='.0f' → for floats without decimals (e.g., 150.0 becomes 150)

* fmt='.1f', '.2f', etc. → for floats with 1 or 2 decimal places (e.g., 123.45)

---

When we do something like:

**img_array[55:165, 365:465]**

We're actually slicing a NumPy array, and in NumPy (as in most image-processing libraries like OpenCV and PIL), the indexing follows the format:

**array[y, x] → [row, column]**

#### So the question comes: Why Y comes before X:

In NumPy arrays, the first index refers to the row (which maps to the Y-axis in image coordinates), and the second index refers to the column (which maps to the X-axis).

So:

* **img_array[55:165, 365:465] means:**

    * Take rows from 55 to 165 → vertical section (Y-axis)

    * Take columns from 365 to 465 → horizontal section (X-axis)
 
#### Important distinction:

* In matplotlib’s plotting functions, like plt.plot(x, y), you specify x first and y second, because that’s standard Cartesian format.

* In image arrays, it's **[y, x]** because:

    * Arrays are 2D: **array[rows, columns]**

    * Images are grids of pixels arranged as **rows (top to bottom)** and **columns (left to right)**
 
But the image we downloaded is stored as a 3D array, typically shaped like:

---

(height, width, channels)

   Y,       X,     (R,G,B)

---

So when you index an image like:

**img_array[55:165, 365:465]**

You're still following this rule:

**img_array[y_range, x_range, :]**

Where:

* y_range → vertical range (rows)

* x_range → horizontal range (columns)

* : → all 3 color channels (RGB)

#### Example:

The image shape is:

**img_array.shape** #Output: (300, 500, 3)

This means:

* 300 pixels in height (Y axis / rows)

* 500 pixels in width (X axis / columns)

* 3 color channels (Red, Green, Blue)

#### Key Insight:

Even though it's a 3D array, the spatial dimensions **(height and width)** are always accessed in **[y, x]** order.
The third dimension just adds color depth, and isn't involved in positioning.
