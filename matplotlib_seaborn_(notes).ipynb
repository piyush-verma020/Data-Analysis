### Chart Legend

In order to differentiate between multiple lines, we can include legend within the graph using the **plt.legend** function 

```python
plt.xlabel('Years')
plt.ylabel('Yield (tonnes per hectare')

plt.title('Crop Yields in Canto')  # "Canto" is an imaginary place somewhere in this cosmos

plt.plot(years, apple)
plt.plot(years, orange)

plt.legend(['Apples', 'Orange'])
plt.show()
```

### Line Markers

We can also show markers for the data point on each line using the **marker** argument of **plt.plot**. Matplotlib supports different varieties of markers
like circle, cross, triangle, diamond e.t.c.

Here is the full list of markers:

**https://matplot.org/3.1.1/api/markers_api.html**

```python
plt.xlabel('Years')
plt.ylabel('Yield (tonnes per hectare')

plt.title('Crop Yields in Canto')  # "Canto" is an imaginary place somewhere in this cosmos

plt.plot(years, apple, marker='D')
plt.plot(years, orange, marker='d')

plt.legend(['Apples', 'Orange'])
plt.show()
```

### Styling lines and markers

The **plt.plot** function supports many argument for styling lines and markers:
* **colour** or **c**: set the colour of the line (https://matplotlib.org/stable/gallery/color/named_colors.html)
* **linestyle** or **ls**: choose between solid or dashed line
* **linewidth** or **lw**: set the width of the line
* **markersize** or **ms**: set the size of the marker
* **markeredgecolour** or **mec**: set the edge colour for markers
* **markerdgewidth** or **mew**: set the edge width for markers
* **markerfacecolour** or **mfc**: set the fill colours for markers
* **alpha**: opacity of the plot


```python
plt.xlabel('Years')
plt.ylabel('Yield (tonnes per hectare')

plt.title('Crop Yields in Canto')  # "Canto" is an imaginary place somewhere in this cosmos

plt.plot(years, apple, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, orange, marker='o', c='r', ls='--', lw=3, ms=10, mec='black',mew=3, alpha=.5)

plt.legend(['Apples', 'Orange'])
plt.show()
```

The **fmt** argument provides a shorthand for specifying the line style, marker and line colour It can be provide as the third argument to **plt.plot**.

The format of the **fmt** argument is: 

**fmt = '[marker][line_style][colour]'**

#### Note:

If no line style is specified in **fmt** argument, then **fmt** will draw only markers.

Well this also alows you to add more colours and make a single plot look more attractive.

```python
plt.xlabel('Years')
plt.ylabel('Yield (tonnes per hectare)')

plt.title('Crop Yields in Canto')  # "Canto" is an imaginary place somewhere in this cosmos

# Plotting Apple yield: blue square markers ('s'), then a black line for trend
plt.plot(years, apple, 'sb')  # Blue square markers for Apple
plt.plot(years, apple, ls='-', c='black', label='Apple')  # Black line for Apple trend

# Plotting Orange yield: red circle markers ('o') with a dashed line
plt.plot(years, orange, 'o--r', label='Orange')  # Red dashed line with circle markers

# Show legend (it uses labels provided in the plot() calls above)
plt.legend()  # Displays the legend with labels 'Apple' and 'Orange'

plt.show()
```
