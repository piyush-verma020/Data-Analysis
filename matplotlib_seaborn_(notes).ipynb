### Chart Legend

In order to differentiate between multiple lines, we can include legend within the graph using the **plt.legend** function 

```python
plt.xlabel('Years')
plt.ylabel('Yield (tonnes per hectare')

plt.title('Crop Yields in Canto')  # "Canto" is an imaginary place somewhere in this cosmos

plt.plot(years, apple)
plt.plot(years, orange)

plt.legend(['Apples', 'Orange'])
plt.show()
```

### Line Markers

We can also show markers for the data point on each line using the **marker** argument of **plt.plot**. Matplotlib supports different varieties of markers
like circle, cross, triangle, diamond e.t.c.

Here is the full list of markers:

**https://matplot.org/3.1.1/api/markers_api.html**

```python
plt.xlabel('Years')
plt.ylabel('Yield (tonnes per hectare')

plt.title('Crop Yields in Canto')  # "Canto" is an imaginary place somewhere in this cosmos

plt.plot(years, apple, marker='D')
plt.plot(years, orange, marker='d')

plt.legend(['Apples', 'Orange'])
plt.show()
```

### Styling lines and markers

The **plt.plot** function supports many argument for styling lines and markers:
* **colour** or **c**: set the colour of the line (https://matplotlib.org/stable/gallery/color/named_colors.html)
* **linestyle** or **ls**: choose between solid or dashed line
* **linewidth** or **lw**: set the width of the line
* **markersize** or **ms**: set the size of the marker
* **markeredgecolour** or **mec**: set the edge colour for markers
* **markerdgewidth** or **mew**: set the edge width for markers
* **markerfacecolour** or **mfc**: set the fill colours for markers
* **alpha**: opacity of the plot


```python
plt.xlabel('Years')
plt.ylabel('Yield (tonnes per hectare')

plt.title('Crop Yields in Canto')  # "Canto" is an imaginary place somewhere in this cosmos

plt.plot(years, apple, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, orange, marker='o', c='r', ls='--', lw=3, ms=10, mec='black',mew=3, alpha=.5)

plt.legend(['Apples', 'Orange'])
plt.show()
```

The **fmt** argument provides a shorthand for specifying the line style, marker and line colour It can be provide as the third argument to **plt.plot**.

The format of the **fmt** argument is: 

**fmt = '[marker][line_style][colour]'**

#### Note:

If no line style is specified in **fmt** argument, then **fmt** will draw only markers.

Well this also alows you to add more colours and make a single plot look more attractive.

```python
plt.xlabel('Years')
plt.ylabel('Yield (tonnes per hectare)')

plt.title('Crop Yields in Canto')  # "Canto" is an imaginary place somewhere in this cosmos

# Plotting Apple yield: blue square markers ('s'), then a black line for trend
plt.plot(years, apple, 'sb')  # Blue square markers for Apple
plt.plot(years, apple, ls='-', c='black', label='Apple')  # Black line for Apple trend

# Plotting Orange yield: red circle markers ('o') with a dashed line
plt.plot(years, orange, 'o--r', label='Orange')  # Red dashed line with circle markers

# Show legend (it uses labels provided in the plot() calls above)
plt.legend()  # Displays the legend with labels 'Apple' and 'Orange'

plt.show()
```

### Scatter Plot

In a sactter plot, the values of 2 variables are plotted as points of a 2-dimentional grid.

Additionally you could also use a third variable to determine
the size or the color of the points.

The **Iris flower dataset** provides sample measurments of sepals and petals for 3 species of flower.

The Iris dataset is included with the seaborn library,and can be loaded in Pandas Data Frame.

```python
# Loading data into Pandas Data Frame
flower_df = sns.load_dataset('iris')
```

```python
flower_df
```

```python
flower_df.species.unique()
```

Let's visualze the relationship between **sepal length** and **sepal width**. Let's create a line chart using **plt.plot()**.

```python
plt.plot(flower_df.sepal_length, flower_df.sepal_width);
```

We can use scatter plot to visualize how sepal length and sepal width vary using the ***scatterplot** function* from seaborn(imported as sns)

### Plotting with Pandas

Seborn has in-built support for Pandas Data Frames. Instead of passing each column as series, 
you can also pass column names using the **data** argument to pass the Data Frame

```python
sns.scatterplot(data=flower_df, x='sepal_length', y='sepal_width');
```

The resulting chart appears visually overwhelming, likely due to the high number of variable combinations in the dataset. 

This complexity makes it challenging to identify any straightforward or meaningful patterns.

### Adding Hues

#### What is hue?

* hue tells Seaborn to separate your data into categories based on a column's values.

* It automatically assigns different colors to each unique value in that column.

#### How to use hue:

##### sns.scatterplot(data=flower_df, x='sepal_length', y='sepal_width', hue='species');

* We are plotting sepal length vs. sepal width.

* The column species has 3 unique values: 'setosa', 'versicolor', 'virginica'.

* hue='species' makes Seaborn:
    * Group the points by species

    * Assign a different color to each species

    * Add a legend so you can understand which color represents which species}
 
#### Visual Effect:
So in our scatterplot, we see:

* Red dots for Setosa

* Blue dots for Versicolor

* Green dots for Virginica
(or other colors — depending on the theme)

Each group is still plotted on the same X-Y space, but with its own color.

```python
sns.scatterplot(data=flower_df, x='sepal_length', y='sepal_width', hue='species');
```

We could also increase the size of the dots by using s = 100 as another argument like:

```python
sns.scatterplot(data=flower_df, x='sepal_length', y='sepal_width', hue='species',s=75);
```

It looks like adding hue makes the plot more informative. we can immediately tell thatthe flower of Setosa species have a smaller sepal length but higher
sepal width, while the opposite is true for the flower of Virginica species.

### Customizing Seaborn Figures

Since seaborn uses matplotlib's plotting functions internally, we can use function like **plt.figure** and **plt.title**
to modify the figure

```python
plt.figure(figsize=(12,6))
plt.title('Sepal Dimension')

sns.scatterplot(data=flower_df, x='sepal_length', y='sepal_width', hue='species',s=75);
```

### Histogram

A histogram represents the distribution of data by forming bins along the range of the data and drawing bars showing
the number of observations that fall in the bin.

As an example, let's visualize how the values of sepal width in the flower data set are distributed.
We can use the **plt.hist** function to create a hostogram.

```python
flower_df.sepal_width
```

```python
flower_df.describe()
```

```python
plt.title('Distribution of Sepal Width')
plt.hist(flower_df.sepal_width);
plt.show()
```

We can imediatly see that the values of sepal width in the range 2.0-4.5, and around 35 values are in the range 2.9-3.1, which seems to be the largest bin.

### Controlling size and numbers of bins

We can control the number and size of each bin just by using the **bins** argument

```python
plt.hist(flower_df.sepal_width,bins=5);
```

```python
import numpy as np

# Specifing the boundaries of each bin
plt.hist(flower_df.sepal_width, bins=np.arange(2,5,0.25));
```

```python
# Bins of unequal size
plt.hist(flower_df.sepal_width, bins=[1,3,4,4.5]);
```

### Multiple Histograms

Similar to line charts, we can draw multiple histogram in a singhe chart. We can reduce the opacity(or alpha) of each histogram, so the bar of one doesn't
hides the other.

```python
flower_df.species.unique()
```

```python
species1 = flower_df[flower_df.species == 'setosa']
```

```python
species2 = flower_df[flower_df['species'] == 'versicolor']
```

```python
species3 = flower_df[flower_df.species == 'virginica']
```

```python
plt.title('Distribution of Sepal Width')
plt.hist(species1.sepal_width, alpha=0.4, bins=np.arange(2,5,0.25),label = 'Setosa');
plt.hist(species2.sepal_width, alpha=0.4, bins=np.arange(2,5,0.25),label = 'Versicolor');
plt.legend()
plt.show()
```

```python
plt.title('Distribution of Sepal Width')
plt.hist([species1.sepal_width, species2.sepal_width, species3.sepal_width], bins=np.arange(2,5,0.25), stacked=True, alpha = 0.75)
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
plt.show()
```

```python
# If not stacked
plt.title('Distribution of Sepal Width')
plt.hist([species1.sepal_width, species2.sepal_width, species3.sepal_width], bins=np.arange(2,5,0.25), stacked=False)
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
plt.show()
```
